```python
import sys
sys.path.append('../')
import lusee
import numpy as np
import matplotlib.pyplot as plt
```

```python
obs = lusee.LObservation(deltaT_sec=60)
S  = lusee.LSatellite()
obsat = lusee.ObservedSatellite (obs,S)
```

```python
fig = plt.figure(figsize=(20,20))
ax = plt.axes(projection='3d')
posxyz = obsat.posxyz
ax.plot3D(posxyz[:2000,0],posxyz[:2000,1],posxyz[:2000,2])
ax.scatter3D(posxyz[:2000:20,0],posxyz[:2000:20,1],posxyz[:2000:20,2])
ax.scatter3D(0,0,0)
```

```python
plt.figure(figsize=(10,10))
obsat.plot_tracks(plt)
plt.xlabel("E-W")
plt.ylabel("N-S")
```
